# Será criado um exemplo de histograma

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/iris.csv')
df.head()
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(df['comprimento_pétala']) # `Só é passado o eixo X
eixo.set_title('Histograma', fontsize=20, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=20)
eixo.grid()
```

```python
# aumentar a quantidade de itens
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(df['comprimento_pétala'], bins=20)  # bins diz quantas barras serão desenhadas no histograma
eixo.set_title('Histograma', fontsize=20, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=20)
eixo.grid()
```

```python
# mudar para horizontal
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(df['comprimento_pétala'], bins=20, orientation='horizontal') 
eixo.set_title('Histograma', fontsize=20, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=20)
eixo.grid()
```

```python
# Visão mais normalizada
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(df['comprimento_pétala'], bins=20, density=True)  # density - Veja Documentação https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html
eixo.set_title('Histograma', fontsize=20, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=20)
eixo.grid()
```

## Calcular a Média e o Desvio Padrão

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

# mu = media  e sigma = desvio padrão
# mean = media  e std= desvio padrão
mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20) 
eixo.set_title('Histograma', fontsize=20, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=20)
eixo.grid()

eixo.annotate('$mu = {0:.2f}$\n$sigma = {1:.2f}$'.format(mu, sigma),
xy = (4.5, 20), fontsize=20) # $$ Formato de formula, para o programa entender que é uma formula
```

```python
# Adicionar um reta media

fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

# mu = media  e sigma = desvio padrão
# mean = media  e std= desvio padrão
mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20) 
eixo.set_title('Histograma', fontsize=20, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=20)
eixo.grid()

# Barra invertidade antes do mu e do sigma, gera os simbolos correspondentes
eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
xy = (4.5, 20), fontsize=20) # $$ Formato de formula, para o programa entender que é uma formula

eixo.axvline(mu, color = 'black', linestyle = '--')
eixo.annotate('média', xy=(mu-1.3, 28), fontsize=20)
```

```python
# Adicionar um reta mediana

fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

# mu = media  e sigma = desvio padrão
# mean = media  e std= desvio padrão
mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20) 
eixo.set_title('Histograma', fontsize=20, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=20)
eixo.grid()

eixo.annotate('$mu = {0:.2f}$\n$sigma = {1:.2f}$'.format(mu, sigma),
xy = (4.5, 20), fontsize=20) # $$ Formato de formula, para o programa entender que é uma formula

eixo.axvline(mu, color = 'black', linestyle = '--')
eixo.annotate('média', xy=(mu-1.3, 28), fontsize=20)

eixo.axvline(df['comprimento_pétala'].median(), color = 'red', linestyle = '--')
eixo.annotate('mediana', xy=(df['comprimento_pétala'].median(), 31), fontsize=20)
```

# Salvar imagem e combinar os histogramas em um único

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

df_ivg = df[df['espécie'] == 'Iris-virginica']

mu, sigma = df_ivg['comprimento_pétala'].mean(), df_ivg['comprimento_pétala'].std()

eixo.hist(df_ivg['comprimento_pétala'], bins = 20)
eixo.grid(True)
eixo.set_title('Histograma Iris Virgínica', fontsize = 15, pad = 10)
eixo.set_xlabel('Comprimento da pétala', fontsize = 15)

eixo.annotate('$\mu = {0:.2f}$\n$\sigma={1:.2f}$'.format(mu, sigma), 
              xy=(6.3, 7), fontsize=20)

eixo.annotate('média', xy=(df_ivg['comprimento_pétala'].mean(), 8), fontsize=20)
eixo.axvline(df_ivg['comprimento_pétala'].mean(), color='k', linestyle='--')

eixo.annotate('mediana', xy=(df_ivg['comprimento_pétala'].median(), 7), fontsize=20, color = 'g')
eixo.axvline(df_ivg['comprimento_pétala'].median(), color='g', linestyle='--')

fig.savefig('data/histograma_iv.png', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

df_iv = df[df['espécie'] == 'Iris-versicolor']

mu, sigma = df_iv['comprimento_pétala'].mean(), df_iv['comprimento_pétala'].std()

eixo.hist(df_iv['comprimento_pétala'], bins = 20)
eixo.grid(True)
eixo.set_title('Histograma Iris Versicolor', fontsize = 15, pad = 10)
eixo.set_xlabel('Comprimento da pétala', fontsize = 15)

eixo.annotate('$\mu = {0:.2f}$\n$\sigma={1:.2f}$'.format(mu, sigma), 
              xy=(4.6, 6), fontsize=20)

eixo.annotate('média', xy=(df_iv['comprimento_pétala'].mean()-0.5, 5.5), fontsize=20)
eixo.axvline(df_iv['comprimento_pétala'].mean(), color='k', linestyle='--')

eixo.annotate('mediana', xy=(df_iv['comprimento_pétala'].median() - 0.7, 6.5), fontsize=20, color = 'g')
eixo.axvline(df_iv['comprimento_pétala'].median(), color='g', linestyle='--')

fig.savefig('data/histograma_iv.png', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

df_is = df[df['espécie'] == 'Iris-setosa']

mu, sigma = df_is['comprimento_pétala'].mean(), df_is['comprimento_pétala'].std()

eixo.hist(df_is['comprimento_pétala'], bins = 10)
eixo.grid(True)
eixo.set_title('Histograma Iris Setosa', fontsize = 15, pad = 10)
eixo.set_xlabel('Comprimento da pétala', fontsize = 15)

eixo.annotate('$\mu = {0:.2f}$\n$\sigma={1:.2f}$'.format(mu, sigma), 
              xy=(1.7, 7), fontsize=20)

eixo.annotate('média', xy=(df_is['comprimento_pétala'].mean()-0.2, 13.5), fontsize=20)
eixo.axvline(df_is['comprimento_pétala'].mean(), color='k', linestyle='--')

eixo.annotate('mediana', xy=(df_is['comprimento_pétala'].median()+0.05, 12), fontsize=20, color = 'g')
eixo.axvline(df_is['comprimento_pétala'].median(), color='g', linestyle='--')

fig.savefig('data/histograma_is.png', bbox_inches='tight')
```

```python
from PIL import Image

largura_imagem, altura_imagem = Image.open('data/histograma_iv.png').size
combinada = Image.new("RGB", (largura_imagem*3, altura_imagem))
x_offset = 0
for imagem in map(Image.open, ['data/histograma_iv.png', 'data/histograma_is.png', 'data/histograma_ivg.png']):
    combinada.paste(imagem, (x_offset, 0))
    x_offset += largura_imagem
combinada.save('combinada.png')
```
