### Convertendo uma coluna para o tipo datetime

```python
df['data'] = pd.to_datetime(df['data'])  # Convertendo para datetime
```

```python
df['data'].tail()  # agora está marcando que é um tipo datetime
```

### Importar o matplot lib para gerar os gráficos

```python
import matplotlib.pyplot as plt
%matplotlib inline  # faz com que os gráficos apareça de forma automatica
```

```python
plt.plot(df['data'], df['temperatura'] )   # data como eixo x e temperatura como eixo y
```

```python
# Mudar tamanho do gráfico
plt.figure(figsize=(15,10))
plt.plot(df['data'], df['temperatura'] ) 
```

```python
# Adicionar titulo e nome dos eixos
plt.figure(figsize=(15,10))
plt.plot(df['data'], df['temperatura'] ) 
plt.title('Temperatura por Ano')
plt.xlabel('Ano')
plt.ylabel('Temperatura K')
```

```python
fig = plt.figure(figsize=(18,8))
# Adicionar um eixo a uma nova figura
eixo = fig.add_axes([0,0,1,1])  # adicionando um eixo com tamanho e medidas
```

```python
# Adicionar o gráfico dentro do eixo vazio
fig = plt.figure(figsize=(18,8))
eixo = fig.add_axes([0,0,1,1]) 
eixo.plot(df['data'], df['temperatura'])   # Todas as mudanças no gráfico pe realizado pelo objeto eixo e não mais pelo plt

eixo.set_title('Temperatura no momento')
```

```python
fig = plt.figure(figsize=(18,8))
eixo = fig.add_axes([0,0,1,1]) 
eixo.plot(df['data'], df['temperatura'])  

eixo.set_title('Temperatura no momento')
eixo.set_ylabel('Temperatura K')
eixo.set_xlabel('Data')
```

```python
fig = plt.figure(figsize=(18,8))
eixo = fig.add_axes([0,0,1,1]) 
eixo.plot(df['data'], df['temperatura'])  

# aumentar o tamanho da fonte
eixo.set_title('Temperatura no momento', fontsize=20)
eixo.set_ylabel('Temperatura K', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)

```

```python
# Adicionar legenda
fig = plt.figure(figsize=(18,8))
eixo = fig.add_axes([0,0,1,1]) 
eixo.plot(df['data'], df['temperatura'])  

# aumentar o tamanho da fonte
eixo.set_title('Temperatura no momento', fontsize=20)
eixo.set_ylabel('Temperatura K', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)  # Lower rigt = inferior direito, pode ser adicionado em diversos locais no gráfico
```

```python
fig = plt.figure(figsize=(18,8))
eixo = fig.add_axes([0,0,1,1]) 
# Mudar a cor do gráfico
eixo.plot(df['data'], df['temperatura'], color = 'red')  


eixo.set_title('Temperatura no momento', fontsize=20)
eixo.set_ylabel('Temperatura K', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)  # Lower rigt = inferior direito, pode ser adicionado em diversos locais no gráfico
```
