# Criar um gráfico de dispersão

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# usando help para o caso de precisar saber mais sobre determinada função ou metodo
help(np.array)
```

```python
df = pd.read_csv('data/iris.csv')
```

```python
df.head()
```

```python
# help(plt.figure)

fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.scatter(df['comprimento_sépala'], df['largura_sépala'])
eixo.set_title('Gráfico de Dispersão', fontsize=25, pad=15)
eixo.set_ylabel('Largura da Sépala', fontsize=20)
eixo.set_xlabel('Comprimento da Sépala', fontsize=20)

# CADA BOLINHA NO GRÁFICO É UM OBJETO, CADA BOLINHA PODE TRAÇAR UM LINHA X E Y
```

```python

fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.scatter(df['comprimento_sépala'], df['largura_sépala'])
eixo.set_title('Gráfico de Dispersão', fontsize=25, pad=15)
eixo.set_ylabel('Largura da Sépala', fontsize=20)
eixo.set_xlabel('Comprimento da Sépala', fontsize=20)

eixo.tick_params(labelsize=15) # aumentar o tamanho da fontes dos numeros
```

```python
# Mudar a cor para cada tipo de sepala
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa': 'red', 'Iris-versicolor':'blue', 'Iris-virginica': 'green' }

# criado um for para ir inserido as cores de acordo com a especie
for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'], color=cores[especie])


eixo.set_title('Gráfico de Dispersão', fontsize=25, pad=15)
eixo.set_ylabel('Largura da Sépala', fontsize=20)
eixo.set_xlabel('Comprimento da Sépala', fontsize=20)

eixo.tick_params(labelsize=15)

eixo.legend(cores, fontsize=20)
```

```python
# Usar marcadores para verificar se tem algum bolinha sobreposta
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa': 'red', 'Iris-versicolor':'blue', 'Iris-virginica': 'green' }
marcadores = {'Iris-setosa': 'x', 'Iris-versicolor':'o', 'Iris-virginica': 'v' }


# criado um for para ir inserido as cores de acordo com a especie
for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'], color=cores[especie],
    marker=marcadores[especie])


eixo.set_title('Gráfico de Dispersão', fontsize=25, pad=15)
eixo.set_ylabel('Largura da Sépala', fontsize=20)
eixo.set_xlabel('Comprimento da Sépala', fontsize=20)

eixo.tick_params(labelsize=15)

eixo.legend(cores, fontsize=20)
```

```python
# Aumentar o tamnho do marcador
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa': 'red', 'Iris-versicolor':'blue', 'Iris-virginica': 'green' }
marcadores = {'Iris-setosa': 'x', 'Iris-versicolor':'o', 'Iris-virginica': 'v' }


# criado um for para ir inserido as cores de acordo com a especie
for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'], color=cores[especie],
    marker=marcadores[especie], s=100)  # s para aumentar o tamanho


eixo.set_title('Gráfico de Dispersão', fontsize=25, pad=15)
eixo.set_ylabel('Largura da Sépala', fontsize=20)
eixo.set_xlabel('Comprimento da Sépala', fontsize=20)

eixo.tick_params(labelsize=15)

eixo.legend(cores, fontsize=20)
```
