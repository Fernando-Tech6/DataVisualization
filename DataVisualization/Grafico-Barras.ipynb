# Criando Gráfico de barras

```python
import pandas as pd
```

```python
df = pd.read_csv('data/monitoramento_tempo.csv')  
```

```python
df.head()
```

```python
df.dtypes
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Mostrar os dias da semana e organizar por ordem de dias
temperatura_por_dia = df.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperatura_por_dia = temperatura_por_dia[nome_dias]  # Eu coloco que a variavel vai passar ter referencia do dia da semana em ordem
temperatura_por_dia
```

```python
figure = plt.figure(figsize=(5,4))
eixo = figure.add_axes([0,0,1,1])  # Criando os tamanhos dos eixos e do gráfico
eixo.bar(temperatura_por_dia, nome_dias)
```

```python
figure = plt.figure(figsize=(5,4))
eixo = figure.add_axes([0,0,1,1])  # Criando os tamanhos dos eixos e do gráfico

indice = range(len(temperatura_por_dia))   # indice para o eixo x, para retornar 1,2,3. len = tamanho 

eixo.bar(indice,temperatura_por_dia)

eixo.set_title('Temperatura por dia', fontsize=20, pad=10)  # pad = distancia do texto com o gráfico
eixo.set_xlabel('Dia da Semana', fontsize = 15)
eixo.set_ylabel('Temperatura', fontsize = 15)
```

```python
# Colocar o indice como os dias da semana
figure = plt.figure(figsize=(5,4))
eixo = figure.add_axes([0,0,1,1])  # Criando os tamanhos dos eixos e do gráfico

indice = range(len(temperatura_por_dia))   # indice para o eixo x, para retornar 1,2,3. len = tamanho 

eixo.bar(indice,temperatura_por_dia)

eixo.set_title('Temperatura por dia', fontsize=20, pad=10)  # pad = distancia do texto com o gráfico
eixo.set_xlabel('Dia da Semana', fontsize = 15)
eixo.set_ylabel('Temperatura', fontsize = 15)

#Colocar o eixo x como strings
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

```python
# Mudar cores do gráfico de barras
figure = plt.figure(figsize=(5,4))
eixo = figure.add_axes([0,0,1,1])  # Criando os tamanhos dos eixos e do gráfico

indice = range(len(temperatura_por_dia))   # indice para o eixo x, para retornar 1,2,3. len = tamanho 

cores = ['black','red', 'blue', 'orange', 'pink', 'green', 'yellow']

eixo.bar(indice,temperatura_por_dia, color = cores, edgecolor='black')

eixo.set_title('Temperatura por dia', fontsize=20, pad=10)  # pad = distancia do texto com o gráfico
eixo.set_xlabel('Dia da Semana', fontsize = 15)
eixo.set_ylabel('Temperatura', fontsize = 15)

#Colocar o eixo x como strings
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```
