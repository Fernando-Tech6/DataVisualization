# Será criado um gráfico de Pizza

```python
import pandas as pd
```

```python
df = pd.read_csv('data/monitoramento_tempo.csv')  
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Mostrar os dias da semana e organizar por ordem de dias
temperatura_por_dia = df.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperatura_por_dia = temperatura_por_dia[nome_dias]  # Eu coloco que a variavel vai passar ter referencia do dia da semana em ordem
temperatura_por_dia
```

```python
figure = plt.figure(figsize=(5,4))
eixo = figure.add_axes([0,0,1,1])

eixo.pie(temperatura_por_dia, labels=temperatura_por_dia.index)  #labels = é o nome das fatias da pizza, que será o index dos dias da semana

eixo.set_title('Temperatura por Dia', fontsize=20)
```

```python
# Colocar porcentagem nas fatias de pizza

figure = plt.figure(figsize=(5,4))
eixo = figure.add_axes([0,0,1,1])

eixo.pie(temperatura_por_dia, labels=temperatura_por_dia.index,
autopct='%.1f%%')  #autopct + o código é para colocar a porcentagem nas fatias de pizza

eixo.set_title('Temperatura por Dia', fontsize=20)
```

```python
# Separar dias especificos

figure = plt.figure(figsize=(5,4))
eixo = figure.add_axes([0,0,1,1])

explodir = [0.1, 0, 0, 0, 0, 0.1 ,0.1]  # 0 = para não habilitar para explodir, no caso está na sequencia dos dias da semana.

eixo.pie(temperatura_por_dia, labels=temperatura_por_dia.index,
autopct='%.1f%%', explode = explodir)  #


eixo.set_title('Temperatura por Dia', fontsize=20)
```
