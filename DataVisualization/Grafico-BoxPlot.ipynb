# Será criado um gráfico de Box-Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/iris.csv')
df.head()
```

```python
 fig = plt.figure(figsize=(5,4))
 eixo = fig.add_axes([0,0,1,1])
 
#plt.boxplot(df['largura_pétala'])   # Uma forma direta sem usar eixos
eixo.boxplot(df['largura_pétala'])  
eixo.set_title('Gráfico de caixa', fontsize=15, pad=10)
eixo.set_xticklabels(['largura_pétala'])
```

```python
 # retirado o especies e criado um boxplot para cada coluna do dataframe
 fig = plt.figure(figsize=(8,5))
 eixo = fig.add_axes([0,0,1,1])
 
# retirar a especie do dataframe
eixo.boxplot(df.drop('espécie', axis=1))  

eixo.set_title('Gráfico de caixa', fontsize=15, pad=10)
eixo.set_xticklabels(df.drop('espécie', axis=1).columns)
```

```python
 # colocar cores para preencher o boxplot
 fig = plt.figure(figsize=(8,5))
 eixo = fig.add_axes([0,0,1,1])
 
# retirar a especie do dataframe
eixo.boxplot(df.drop('espécie', axis=1), patch_artist= True)  

eixo.set_title('Gráfico de caixa', fontsize=15, pad=10)
eixo.set_xticklabels(df.drop('espécie', axis=1).columns)
```

```python
 # trocar marcadores e mudar cores

 fig = plt.figure(figsize=(8,5))
 eixo = fig.add_axes([0,0,1,1])
 
# retirar a especie do dataframe
caixas = eixo.boxplot(df.drop('espécie', axis=1), patch_artist= True)  

eixo.set_title('Gráfico de caixa', fontsize=15, pad=10)
eixo.set_xticklabels(df.drop('espécie', axis=1).columns)

cores = ['red', 'blue', 'orange', 'green']

for caixa, cor in zip(caixas['boxes'], cores):
    caixa.set(color = cor)

for outlier in caixas['fliers']:
    outlier.set(marker='x', markersize=8)
```
